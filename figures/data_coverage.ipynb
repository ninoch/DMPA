```python
import pickle
import igraph
import scipy.stats
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib as mpl
import matplotlib.pyplot as plt
import matplotlib.lines as mlines
from matplotlib.patches import Rectangle
from matplotlib.collections import PatchCollection
plt.style.use('seaborn-white')
plt.rc('font', size=70) 
```

```python
fields = ['economics', 'management', 'psychology', 'political_science', 'aps', 'cs']
fields = ['management', 'aps', 'psychology', 'political_science', 'economics', 'cs']
```

```python
cp = sns.color_palette("Set1")
field_color = {
    'economics': cp[0],
    'management': cp[1],
    'psychology': cp[7],
    'aps': cp[3],
    'political_science': cp[2],
    'cs': cp[4]
}
field_label = {
    'economics': 'Economics',
    'management': 'Management',
    'psychology': 'Psychology',
    'aps': 'Physics (APS)',
    'political_science': 'Political Science',
    'cs': 'Computer Science'
}
field_marker = {
    'economics': 'o',
    'management': 's',
    'psychology': '^',
    'aps': '>',
    'political_science': 'd',
    'cs': '<'
    
}
```

```python
for field in fields:
    years, known = pickle.load(open("../params/coverage_{}.pkl".format(field), "rb"))
    plt.plot(years, known, label = field_label[field], color = field_color[field], marker = field_marker[field], markersize=9)
    
    print (years)

years = range(1990, 2021)
plt.xticks(years, years, rotation=90, fontsize=20)
ytck = np.arange(0.05, 1.05, 0.05)
plt.yticks(ytck, ["{0:.2f}".format(y) for y in ytck], fontsize=20)
plt.ylabel('Ratio of the authors with known gender', fontsize=16)
plt.gcf().set_size_inches(13, 10, forward=True)
plt.legend(prop={'size': 20}, facecolor='white', framealpha=1, frameon=True)
plt.savefig("coverage.pdf", bbox_inches='tight')
plt.show()
```

```python
for field in fields:
    years, known = pickle.load(open("../params/coverage_aff_{}.pkl".format(field), "rb"))
    plt.plot(years, known, label = field_label[field], markeredgecolor = field_color[field], color=field_color[field], marker = field_marker[field], markersize=9, markerfacecolor='none')

    print (years)

years = range(1990, 2021)
plt.xticks(years, years, rotation=90, fontsize=20)
ytck = np.arange(0.05, 1.05, 0.05)
plt.yticks(ytck, ["{0:.2f}".format(y) for y in ytck], fontsize=20)
plt.ylabel('Ratio of the authors with known affiliation', fontsize=16)
plt.gcf().set_size_inches(13, 10, forward=True)
# plt.legend(prop={'size': 20}, facecolor='white', framealpha=1, frameon=True)
plt.savefig("coverage_aff.pdf", bbox_inches='tight')
plt.show()
```

```python
leg_lines = []

for field in fields:
    years, known = pickle.load(open("../params/coverage_{}.pkl".format(field), "rb"))
    plt.plot(years, known, label = field_label[field], color = field_color[field], marker = field_marker[field], markersize=9)
    
    years, known = pickle.load(open("../params/coverage_aff_{}.pkl".format(field), "rb"))
    plt.plot(years, known, label = field_label[field], markeredgecolor = field_color[field], color=field_color[field], marker = field_marker[field], markersize=9, markerfacecolor='none')

    leg_lines.append(mlines.Line2D([], [], color=field_color[field], marker=field_marker[field], markersize=7, label="{}: Gender".format(field_label[field]), linestyle='None'))

for field in fields:
    leg_lines.append(mlines.Line2D([], [], markeredgecolor=field_color[field], marker=field_marker[field], markersize=7, label="Affiliation".format(field_label[field]), linestyle='None', markerfacecolor='none'))

years = range(1990, 2021)
plt.xticks(years, years, rotation=90, fontsize=20)
ytck = np.arange(0.0, 1.05, 0.1)
plt.yticks(ytck, ["{0:.1f}".format(y) for y in ytck], fontsize=20)
plt.ylabel('Ratio of the authors with known attribute', fontsize=20)
plt.gcf().set_size_inches(13, 10, forward=True)
plt.legend(handles=leg_lines, prop={'size': 16}, facecolor='white', ncol=2, framealpha=1, frameon=True, bbox_to_anchor=(0.5, 0.53), edgecolor = 'none', loc='center')
plt.savefig("coverage_all.pdf", bbox_inches='tight')
plt.show()
```
